```python
#Reading data for part 1A and storing the data in dataframe "df"
df=pd.read_csv(r"C:\Users\shashidhar\Desktop\Data Science\ab40c054-5031-4376-b52e-9813e776f65e.csv")
df
```

```python
#Getting concise summary about the dataframe to find out how many null values are present in each column, how many columns the dataframe has etc.
df.info()
```

```python
#Deleting the rows corresponding to the year 2016-17 as the values are mainly Nan and it is not useful to us.
df.drop(df[df.Duration=="2016-17"].index,inplace=True)

#Cross-checking if the dataframe still has any rows corresponding to the duration of 2016-17
print(df["Duration"].value_counts())

#Checking the shape of the dataframe after deleting the rows corresponding to 2016-17.
print(df.shape)
```

```python
#Deleting West Bengal column as all values are null
del df["West Bengal1"]

#Checking shape of the dataframe after deleting the column corresponding to West Bengal
df.shape
```

```python
#Taking ""% Growth over previous year" rows for the years 2013-14,2014-15 and 2015-16 for all states and storing it in a new dataframe df1
df1 = df.iloc[6:10,1:]

#Transposing the above dataframe in order  to calculate average growth
df1 = df1.T

#Making the Index into a separate column.
df1.reset_index(level=0, inplace=True)

#Renaming the columns
df1.columns=["States","2013-14","2014-15","2015-16"]

#Dropping the first row as it is repeated as column header.
df1.drop(df1.index[0],inplace=True)

#Calculating the average growth percentage for each state. We will be ignoring the null values in some of the columns. 
df1["mean"]= df1.iloc[:,1:].mean(axis=1)

#Sorting the dataframe based on the "mean" column
df1=df1.sort_values(by="mean",ascending=False)
df1= df1.reset_index(drop=True)
df1
```

```python
#Plotting a bar graph using seaborn to visualize the growth % of all states in the given duration.
plt.figure(figsize=(20, 6))
g=sns.barplot(x='States', y = 'mean',errwidth=3.0,data = df1)
plt.xticks(rotation=90)
for p in g.patches:
    g.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width() / 2., p.get_height()),
           ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.xlabel("STATES")
plt.ylabel("AVERAGE GROWTH %")
plt.title("AVERAGE % GROWTH FOR 2013-14, 2014-15 & 2015-16")
plt.show()
```

From the graph we can see that the average growth percentage is Highest for Mizoram and the least for Goa during the years 2013-14, 2014-15 and 2015-15. Meghalaya has a very low average growth rate as well at 7%. 

```python
#Creating a dataframe from df1_T to compare my state(Karnataka) GSDP vs All_India GDP
df_karnataka = df1.iloc[df1.index[[5,25]]]
df_karnataka

plt.figure(figsize=(6, 6))
g=sns.barplot(x='States', y = 'mean',errwidth=3.0,data = df_karnataka)
plt.xticks(rotation=90)
for p in g.patches:
    g.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width() / 2., p.get_height()),
           ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.xlabel("STATES")
plt.ylabel("AVERAGE GROWTH %")
plt.title("COMPARISON BETWEEN AVERAGE GROWTH RATE OF KARNATAKA NATIONAL AND NATIONAL AVERAGE FROM 2013-2016")
plt.show()
```

The average growth rate of Karnataka during the period of 2013 to 2016 is 14.1%. The National average during this period is 11.2%. From the below graph, we can infer that the average growth rate of Karnataka is higher than the national average by 2.9%.

The total GSDP in the year 2015-16 vs states bar plot shows that States like Tamil Nadu, Uttar Pradesh and Karnataka are

```python
#Reading all files into a single dataframe named "m_df"

import glob

pd.options.mode.chained_assignment = None 
#pd.options.display.float_format = '{:,.0f}'.format

folder = r"C:\Users\shashidhar\Desktop\GDP assignment files\*.csv"
all_files= glob.glob(folder)

col_1=["Item"]
col_2=["2014-15"]

m_c=0
m_df=pd.DataFrame()
for csv in all_files:
    stat=csv.split("-")[1]
    if m_c<1:
        m_df=pd.read_csv(csv,encoding="ISO-8859-1",skipinitialspace=True,usecols=col_1)
        m_c +=1
    em_df=pd.read_csv(csv,encoding="ISO-8859-1",skipinitialspace=True,usecols=col_2)
    em_df.columns=em_df.columns.map(lambda x:stat)
    m_df=pd.merge(m_df,em_df,right_index=True,left_index=True)
        
m_df=m_df.T
m_df.reset_index(level=0, inplace=True)
new_header = m_df.iloc[0] 
m_df = m_df[1:]
m_df.columns = new_header
m_df.rename(columns = {"Item": "States"},inplace = True) 
m_df
```

```python
#We create a duplicate dataframe to perform all transformations and cleaning operations so that we can use the above dataframe in Part 2.
partb_df=m_df

#Deleting all Union Territories.
partb_df.drop(partb_df.index[[0,5,7,23]],inplace=True)
partb_df
```

```python
#Sorting the dataframe in descending order by the "Per Capita GSDP" column to plot graph.Sorting helps us identify top and bottom 5 easily.
partb_df=partb_df.sort_values(by="Per Capita GSDP (Rs.)",ascending=False)

#Plotting the GDP per capita for all the states
plt.figure(figsize=(25, 9))
g=sns.barplot(x='States', y = 'Per Capita GSDP (Rs.)',errwidth=3.0,data = partb_df)
plt.xticks(rotation=90)
for p in g.patches:
    g.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width() / 2., p.get_height()),
           ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.title("Per Capita GSDP (Rs.) for the year 2014-15")
plt.show()
```

Based on GSDP Per Capita, we can infer from the above graph that Goa is the highest State with a value of Rs. 2,71,793 with Sikkim falling closely behind with a GSDP Per Capita value of Rs. 2,40,274 and Bihar  is the lowest state with a value of Rs. 33,954.

```python
#Creating a subset of the main dataframe with only the required columns to Plot the percentage contribution of the primary, secondary and tertiary sectors versus the total GDP for all the states.
df3= partb_df[['States','Primary','Secondary','Tertiary','Gross State Domestic Product']]
#Sorting the dataframe in descending order by the "Gross State Domestic Product" column to obtain a meaningful plot.
df3=df3.sort_values(by="Gross State Domestic Product",ascending=False)
df3
```

```python
#Creating three new columns with calculated percentage values for primary, secondary and tertiary sectors respectively
df3['%Primary'] = (((df3['Primary'])/(df3['Gross State Domestic Product']))*100)
df3['%Secondary'] = (((df3['Secondary'])/(df3['Gross State Domestic Product']))*100)
df3['%Tertiary'] = (((df3['Tertiary'])/(df3['Gross State Domestic Product']))*100)
df3
```

```python
#Creating Stacked bar to see which of the three sectors (Primary, Secondary or Tertiary) contributes the most to the total GDP for each state.
barPrimary=df3.iloc[:,5]
barSecondary=df3.iloc[:,6]
barTertiary=df3.iloc[:,7]
bar_states=df3.iloc[:,0]


plt.figure(figsize=(16,7))
graphPrimary=plt.bar(x=bar_states,height=barPrimary,width=0.35,label="Primary%")
graphSecondary=plt.bar(x=bar_states,height=barSecondary,width=0.35,bottom=barPrimary,label="Secondary%")
graphTertiary=plt.bar(x=bar_states,height=barTertiary,width=0.35,bottom=barSecondary+barPrimary,label="Tertiary%")

plt.xticks(rotation=90)
plt.xlabel("States")
plt.ylabel("Percentage %")
plt.title("PERCENTAGE CONTRIBUTION OF PRIMARY, SECONDARY AND TERTIARY SECTORS")

plt.legend(loc="upper left",bbox_to_anchor=(1, 0.5))
plt.show()
```

A Stacked bar chart shows the contribution of Primary, Secondary and Tertiary Sector to the Total GSDP for every state. We can see that for most States (e.g. Maharashtra, Manipur, Mizoram, Uttar Pradesh, Karnataka, Tamil Nadu, Rajasthan, Andhra Pradesh, Kerala, Telangana, Haryana, Punjab, Odisha, Jharkhand, Assam, Jammu Kashmir, Goa, Tripura and Nagaland), the highest contribution is through the Tertiary sectors.
